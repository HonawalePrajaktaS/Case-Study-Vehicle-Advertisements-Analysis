# drop the outliers --> if it is neccessory

```python
numeric_df= df.select_dtypes(include='number')
q1=numeric_df.quantile(0.25)
q3=numeric_df.quantile(0.75)
IQR=q3-q1
LF=q1-1.5*IQR
UF=q3+1.5*IQR

df_filtered = numeric_df[(numeric_df >= LF) | (numeric_df <= UF)]
df= df_filtered
```

The sns.heatmap(df.corr(), cmap="BrBG", annot=True) --> command is used to create a heatmap of the correlation matrix of a Pandas DataFrame df.

df.corr()     --> This method calculates the pairwise correlation of all columns in the DataFrame df. 
              --> The result is a square matrix where the entry at row i and column j represents the correlation between                      the i-th and j-th columns.
sns.heatmap() --> This is a function from the Seaborn library that creates a heatmap from a 2D array.
cmap="BrBG"   --> This specifies the color map to use for the heatmap. BrBG is a diverging color map that ranges from blue                    (negative correlations) to red (positive correlations) with a neutral white color in the middle (no                          correlation).
annot=True    --> This parameter adds numerical annotations to each cell in the heatmap, showing the exact correlation                        value. If annot=False, the heatmap will only display the colors.
